```python
Länderkürzel = "Ajbals - Österreich,BayZR - Deutschland,GB - Deutschland,Digital - Recht - Deutschland,(Kein Kürzel)EJELS - Österreich,EALR - Österreich,FoR - Deutschland,GJLE - Deutschland,GoJIL - Deutschland,GRZ - Deutschland,HFR - Deutschland,JIPITEC - Deutschland,KrimOJ - Deutschland,JLL - Niederlande,(englische Artikel)OdW - Deutschland,MIR - Deutschland,MIP - Deutschland,ZRI - Deutschland,Rg - Deutschland,SPWR - Liechtenstein,SutdZR-WissOn - Deutschland,(Achtung, wenn man Studentisches Forum für Recht und Gesellschaft googelt, findet man eine schweizer Zeitschrift. DIese ist allerdings aus Heidelberg)VLR - Österreich,EuZ - Schweiz,ZJS - Deutschland,ZIS - Deutschland,(Zeitschrift für internationale Strafrechtsdogmatik - hat leider auch ZJS als Kürzel im Datensatz eingetragen)ZVR - Deutschland,(steht nirgendow exklusiv aber klingt so)Zeitschrift für freiberufliche Künstler und Publizisten im Recht - Deutschland (Kein Kürzel)"

# Remove parts of the string within brackets
output_string = ''
inside_brackets = False

for char in Länderkürzel:
    if char == '(':
        inside_brackets = True
    elif char == ')':
        inside_brackets = False
    elif not inside_brackets:
        output_string += char

output_string

output_list = output_string.split(',')

final_list = []
for entry in output_list:
  kv = entry.split(' - ')
  final_list.append(kv)

final_list[3] = ['Digital - Recht', 'Deutschland']

final_list
```

Jetzt die Ländernamen den einzelnen Einträgen zuordnen

```python
import numpy as np

df['Land'] = '-'

for entry in final_list:
  df.loc[df['Zeitschrift'].str.contains(entry[0]) | df['Zeitschriftkürzel'].str.contains(entry[0]), 'Land'] = entry[1]

df.head()
```
