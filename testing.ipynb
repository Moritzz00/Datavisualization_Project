```python
import pandas as pd
```

```python
file = "2023-12-07 Datensatz.csv"
df = pd.read_csv(file, encoding="UTF8")
```

```python
df.head()
```

```python
df.info()
```

```python
df.dropna(axis=1, how='all', inplace = True)
df.info()
```

```python
PDF_Gesamtausgabe_Link_Druck = df[df['PDF_Gesamtausgabe_Link_Druck'].notna()]
PDF_Gesamtausgabe_Link_Druck
```

```python
Mit_Autor = df[df['Autor ohne Grad'].notna() | df['Autor mit Grad'].notna()]
Mit_Autor
```

```python
author_df = df['Autor ohne Grad'].str.split(', ', expand=True).stack().reset_index(level=1, drop=True).rename('Autor ohne Grad')
author_counts = author_df.value_counts().sort_values(ascending=False)

print("\nAutor Contribution Counts (descending order):")
print(author_counts)
```

```python
df['all_authors'] = df['Autor ohne Grad'].str.cat(df['Autor mit Grad'], sep=', ', na_rep='')
author_df = df['all_authors'].str.split(', ', expand=True).stack().reset_index(level=1, drop=True).rename('author')

author_counts = author_df.value_counts().sort_values(ascending=False)

print("\nAuthor Contribution Counts (descending order):")
print(author_counts)
```

```python
unique_abbr = df['Zeitschriftkürzel'].unique()

print('\nUnique Abbreviations:')
print(unique_abbr, len(unique_abbr))
```
